# Day 3 - Data Cleaning Fundamentals

## Dataset
Automobile Dataset

### Objectives

- Find missing values
- Fill missing values
- Remove missing values
- Find duplicate rows
- Remove duplicates
- Check data types
- Convert data types

```python
import pandas as pd
```

```python
df = pd.read_csv("AUTOMOBILE DATASET.csv")
```

```python
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
missing_percentage
```

```python
df_filled = df.fillna("Unknown")
```

```python
df_filled.head()
```

```python
df_dropna = df.dropna()
```

```python
df_dropna.shape
```

```python
df.duplicated()
```

```python
df.duplicated().sum()
```

```python
df_no_duplicates = df.drop_duplicates()
```

```python
df_no_duplicates.shape
```

```python
df.dtypes
```

```python
df.info()
```

```python
df["price"] = df["price"].astype(float)
```

```python
df["price"] = pd.to_numeric(df["price"], errors="coerce")
```

```python
df["price"].dtype
```

```python
df.to_csv("../data/automobile_cleaned.csv", index=False)
```
