# The Z-Score Method

One common method of detecting outliers is the **Z-Score Method**.

The Z-Score measures how many Standard Deviations a value is away from the Mean.

Formula:

z = (Value − Mean) / Standard Deviation

Interpretation:

| Z-Score | Meaning |
|----------|---------|
| 0 | Exactly at the mean |
| 1 | One standard deviation above the mean |
| -1 | One standard deviation below the mean |
| 3 | Very unusual value |
| -3 | Very unusual value |

A common rule is:

If

|Z| > 3

then the value is considered a potential outlier.

# Finding Outliers Using Standard Deviation

Now that we understand the Mean and Standard Deviation, we can identify potential outliers.

The process is:

1. Select one numeric column.
2. Calculate its mean.
3. Calculate its standard deviation.
4. Define the lower limit:
   
   Mean − (3 × Standard Deviation)

5. Define the upper limit:

   Mean + (3 × Standard Deviation)

6. Any value outside these limits is considered a potential outlier.

This is called the **3 Standard Deviation Rule**.

```python
column = "price"
```

```python
mean = df[column].mean()
```

# Observation

The variable **mean** now stores the average price of all automobiles in the dataset.

This value will be used to determine the normal range of prices.

```python
mean
```

```python
std = df[column].std()
```

# Observation

The variable **std** stores the Standard Deviation of the Price column.

It tells us how much the prices vary from the average price.

```python
std
```

# Three Standard Deviation Rule

Most normal values fall inside

Mean ± 3 Standard Deviations

Anything outside that range deserves investigation.

```python
lower_limit = mean - (3 * std)
```

```python
upper_limit = mean + (3 * std)
```

```python
print(lower_limit)
print(upper_limit)
```

```python
outliers = df[
    (df[column] < lower_limit) |
    (df[column] > upper_limit)
]
```

```python
df[column]
```

# Observation

The output contains only rows that fall outside the normal range.

If the DataFrame is empty, it means no potential outliers were detected for the selected column using the ±3 Standard Deviation Rule.

An empty result is not an error—it simply indicates that all values fall within the expected range.

```python
outliers
```

```python
print("Number of Outliers:", len(outliers))
```

# Automating Outlier Detection

Previously, we checked only one column (Price).

Now we will automate the process.

Instead of manually selecting a column, Python will:

1. Detect every numeric column.
2. Calculate the mean.
3. Calculate the standard deviation.
4. Calculate the lower and upper limits.
5. Find outliers.
6. Print a report for every numeric column.

This makes the code reusable for any dataset.

```python
numeric_columns = df.select_dtypes(include="number").columns
```

```python
numeric_columns
```

```python
for column in numeric_columns:
    print(column)
```

# Understanding the Loop

For every numeric column, Python performs the following steps:

1. Calculates the Mean.
2. Calculates the Standard Deviation.
3. Computes the Lower Limit.
4. Computes the Upper Limit.
5. Finds values outside these limits.
6. Counts the number of outliers.
7. Prints the result.

This process is repeated automatically for every numeric column.

for column in numeric_columns:

    mean = df[column].mean()

    std = df[column].std()

    lower_limit = mean - (3 * std)

    upper_limit = mean + (3 * std)

    outliers = df[
        (df[column] < lower_limit) |
        (df[column] > upper_limit)
    ]

    print(column)

    print("Outliers:", len(outliers))

    print("-" * 40)

# Summary of Automated Outlier Detection

In this section, we automated the process of detecting outliers for every numeric column.

The steps performed were:

1. Identify all numeric columns.
2. Loop through each numeric column.
3. Calculate the Mean.
4. Calculate the Standard Deviation.
5. Calculate the Lower and Upper Limits.
6. Detect values outside the expected range.
7. Count the number of outliers.
8. Print a formatted report.

This approach is reusable and can work with any dataset containing numeric columns.

```python
for column in numeric_columns:

    mean = df[column].mean()

    std = df[column].std()

    lower_limit = mean - (3 * std)

    upper_limit = mean + (3 * std)

    outliers = df[
        (df[column] < lower_limit) |
        (df[column] > upper_limit)
    ]

    print("=" * 50)

    print("Column:", column)

    print("Mean:", round(mean, 2))

    print("Standard Deviation:", round(std, 2))

    print("Lower Limit:", round(lower_limit, 2))

    print("Upper Limit:", round(upper_limit, 2))

    print("Number of Outliers:", len(outliers))
```
